### Different ways to create DataFrame

#### Using Dictionary

```python
weather_data ={
    'day': ['1/1/2022','1/2/2022','1/3/2022','1/4/2022','1/5/2022','1/6/2022'],
    'temperature': [32,35,33,32,34,36],
    'windspeed': [6,7,8,7,5,4],
    'event': ['Rain','Sunny','Snow','Rain','Sunny','Snow']
    
}

df=pd.DataFrame(weather_data)
df
```

#### Using Excel

```python
import pandas as pd
df=pd.read_excel('weather_data.xlsx',)
df
```

#### Using list and tuple

```python
weather_data=[
    ('1/1/2022',32,6,'Rain'),
    ('1/2/2022',35,6,'Snow'),
    ('1/3/2022',30,6,'Sunny')
]
df=pd.DataFrame(weather_data)
df
```

```python
weather_data=[
    ('1/1/2022',32,6,'Rain'),
    ('1/2/2022',35,6,'Snow'),
    ('1/3/2022',30,6,'Sunny')
]
df=pd.DataFrame(weather_data, columns=['day','temperature','windspeed','event'])
df
```

#### Using list of dictionary

```python
weather_data=[
    {'day':'1/1/2022', 'temperature':32, 'windspeed':6, 'event':'Rain'},
    {'day':'1/2/2022', 'temperature':33, 'windspeed':7, 'event':'Sunny'},
    {'day':'1/3/2022', 'temperature':31, 'windspeed':6, 'event':'Snow'}
]
df=pd.DataFrame(weather_data)
df
```

## Handling Missing values

```python
df=pd.read_csv('weather_data2.csv',parse_dates=['day'])
df
```

```python
df.set_index('day',inplace=True)
```

```python
df
```

## fillna

Fill not available

```python
new_df=df.fillna(0)
new_df
```

```python
new_df=df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no event'
})
new_df
```

## Forward fill

```python
new_df=df.fillna(method='ffill')
new_df
```

## Backward Fill

```python
new_df=df.fillna(method='bfill')
new_df
```

## Column Wise Fill

```python
new_df=df.fillna(method='bfill',axis='columns')
new_df
```

```python
new_df=df.fillna(method='ffill',axis='columns')
new_df
```

```python
df
```

## dropna

```python
new_df=df.dropna()
new_df
```

```python
new_df=df.dropna(how='all')
new_df
```

```python
new_df=df.dropna(thresh=2)
new_df
```
